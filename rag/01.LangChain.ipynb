```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
    model="openai/gpt-oss-20b:free",
     api_key=os.environ["API_KEY"],
      base_url="https://openrouter.ai/api/v1")

messages = [
    ("system", "사용자가 입력한 문장을 영어로 번역해"),
    ("human", "오늘은 피자를 먹어야지")
]

result = llm.invoke(messages)

result
```
